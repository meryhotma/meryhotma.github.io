# Step 2
Calculating mean, median and mode of Export based on CIF Value üí∏


```python
# mean
mean_export = df["cifvalue"].mean()
# Round to 2 decimal places
mean_export = round(mean_export, 2)

# median
median_export = df["cifvalue"].median()
# Round to 2 decimal places
median_export = round(median_export, 2)

# mode
mode_export = df["cifvalue"].mode()
# Round to 2 decimal places
mode_export = round(mode_export, 2)

print("Mean of CIF Value-Export: ", mean_export)
print("Median of CIF Value-Export: ", median_export)
print("Mode of CIF Value-Export:", mode_export)
```

note on mode (using pandas):
CIF value in this dataset is a monetary measure of export value in USD

the pandas.series.mode() returns all values that occur with the highest frequency,

when every value appears exactly once, the highest freq is 1 for every values

hence, when we calculated the mode, pandaas  returns the entire column.



# The Hard Way üí™üèº
calculating mean, median, mode without using pandas, the statistics module, a spreadsheet program, etc üî¢

```python
# import necessary module

import csv as csv
# open the file with "with open"
# Reads the file row by row, using commas as separators

values = []

with open("Trade.csv", "r", encoding="latin1") as file:
    reader = csv.reader(file, delimiter=",")

    header = next(reader)
    header = [h.strip().lower() for h in header]

    cif_value = header.index("cifvalue")

    for row in reader:
        if len(row) < cif_value + 1:
            continue
        try:
            values.append(float(row[cif_value + 1]))
        except ValueError:
            continue

print(f"Loaded {len(values)} cifvalue entries")
```

```python
# calculate mean, median and mode without pandas and statistics

# mean
mean_manual = sum(values) / len(values)
mean_manual = round(mean_manual, 2)
mean_manual

# median
sorted_value = sorted(values)
n = len(sorted_value)
if n % 2 == 1:
    median_manual = sorted_value[n // 2]
else:
    median_manual = (sorted_value[n // 2 - 1] + sorted_value[n // 2]) / 2
median_manual = round(median_manual, 2)
median_manual

# mode
counts = {}
for i in sorted_value:
    counts[i] = counts.get(i, 0) + 1
    # find the most frequent value from counts
max_counts = max(counts.values())
mode_manual = [val for val, freq in counts.items() if freq == max_counts]
mode_manual = [round(m, 2) for m in mode_manual]

mode_manual
# mode, calculated manually is presented in a list format.

print("Mean of CIF Value-Export, manually calculated: ", mean_manual)
print("Median of CIF Value-Export, manually calculated: ", median_manual)
print("Mode of CIF Value-Export, manually calculated:", mode_manual)
```

note on mode (calculated "manually):

to make mode in manual calculation consistent with pandas.series.mode(), 

the mode is computed as a list of all values that appears at highest frequency

## Step 1
aggregate CIF Value by month (use refmonth) using pandas (allowed based on the assignment prompt)

```python
df["refMonth"].unique()

# refmonth format is YYYYMM
# Derive month value from refmonth and change to interger
df["month_num"] = df["refMonth"].astype(str).str[-2:].astype(int)

df["month_num"]
```

```python
# group data by month
monthly = df.groupby("month_num")["cifvalue"].sum().sort_index()

months = list(monthly.index)
values = list(monthly.values)

# change month label to make the viz easier to read
month_label = {
    1: "Jan",
    2: "Feb",
    3: "Mar",
    4: "Apr",
    5: "May",
    6: "Jun",
    7: "Jul",
    8: "Aug",
    9: "Sep",
    10: "Oct",
    11: "Nov",
    12: "Dec",
}

labels = [month_label[m] for m in months]

labels
```

## Step 2
Creat the viz, yay!

```python
# time series comparison
def ascii_time_series(labels, values, width=50, symbol="‚ñà"):
    # Visualize CIFVALUE per month (USD) using only standard Python.
    max_val = max(values)
    scale = max_val / width if max_val else 1

    print("\nCIFVALUE by Month (2024, USD)")
    print("‚îÄ" * 75)
    print(f"{'Month':<10}{'Trade Value (USD)':>25}")
    print("‚îÄ" * 75)

    for lab, v in zip(labels, values):
        bar_len = int(v / scale)
        bar = symbol * bar_len
        print(f"{lab:<10}| {bar:<50} {v:,.2f}")

    print("‚îÄ" * 75)
    print("Axis: X = CIFVALUE (USD, scaled), Y = Month (Jan‚ÄìDec 2024)")
    print("Note: CIFVALUE = Cost, Insurance, and Freight value of exports (USD).\n")


ascii_time_series(labels, values)
```

```python
def sparkline(nums):
    # Draw compact trend line using Unicode blocks.
    ticks = "‚ñÅ‚ñÇ‚ñÉ‚ñÑ‚ñÖ‚ñÜ‚ñá‚ñà"
    if not nums:
        return ""
    lo, hi = min(nums), max(nums)
    if hi == lo:
        return ticks[0] * len(nums)
    return "".join(ticks[int((x - lo) / (hi - lo) * (len(ticks) - 1))] for x in nums)


print("Monthly CIFVALUE trend (2024):")
print("Months:", labels[0], "‚Üí", labels[-1])
print("Trend: ", sparkline(values))
```
